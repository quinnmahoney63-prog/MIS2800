```python
%%sql
CREATE TABLE Marketing_Brands (
  brand_id INTEGER PRIMARY KEY,
  brand_name TEXT,
  brand_type TEXT,
  brand_budget INTEGER
) STRICT;
```

```python
%%sql
INSERT INTO Marketing_Brands (brand_id, brand_name, brand_type, brand_budget)
VALUES
  (1, 'Doritos', 'Snacks', 7000000),
  (2, 'AWS', 'Technology', 15000000),
  (3, 'Pepsi', 'Beverages', 2000000),
  (4, 'He_Gets_Us', 'Religious', 1000000),
  (5, 'Dr_Pepper', 'Beverages', 5000000),
  (6, 'IBM', 'Technology', 100000),
  (7, 'Google', 'Technology', 8000000);
```

```python
%%sql
-- Instructions: Select all columns from the Marketing_Brands table
SELECT * FROM Marketing_Brands;
```

```python
%%sql
-- Instructions: Display only the brand names and brand budgets
SELECT brand_name, brand_budget
FROM Marketing_Brands;
```

```python
%%sql
-- Instructions: Find the brands where the brand type is Religious
SELECT *
FROM Marketing_Brands
WHERE brand_type = 'Religious';
```

```python
%%sql
-- Instructions: Display the brand names that have a budget greater than 5000000
SELECT brand_name
FROM Marketing_Brands
WHERE brand_budget > 5000000;
```

```python
%%sql
-- Instructions: Run this in SQLite3 in Colab without errors
SELECT *
FROM Marketing_Brands
WHERE brand_budget > 7000000;
```

```python
%%sql
-- Instructions: Find the brands where the brand type is Snacks or Beverages
SELECT *
FROM Marketing_Brands
WHERE brand_type = 'Snacks' OR brand_type = 'Beverages';
```

```python
%%sql
-- Instructions: Sort by brand budget in DESC order. Display only brand name and budget.
SELECT brand_name, brand_budget
FROM Marketing_Brands
ORDER BY brand_budget DESC;
```
