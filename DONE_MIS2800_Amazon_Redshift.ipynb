2. Speed to Insight: How does the architecture of a "Leader Node" and "Compute Nodes" change how quickly a Business Intelligence report is generated?  https://docs.aws.amazon.com/redshift/latest/dg/c_high_level_system_architecture.html


## YOUR ANSWERS TO PART 2 BELOW


Question 1: Columnar storage is more important for analysts because it allows Redshift to store and read data by column instead of by row. Since analytical queries usually focus on a small number of columns and scan large datasets, columnar storage reduces the amount of data read and improves query performance. This provides faster analysis and reporting, even if the data is not perfectly normalized.

Question 2: The Leader Node receives and optimizes queries, then distributes the work to multiple Compute Nodes. The Compute Nodes process the data in parallel, each working on a portion of the dataset at the same time. This parallel processing significantly reduces the time required to generate Business Intelligence reports.
