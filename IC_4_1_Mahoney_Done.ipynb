```python
%%sql
-- Create a table with STRICT types
CREATE TABLE IF NOT EXISTS sports_gear (
    gear_id INTEGER PRIMARY KEY,
    gear_name TEXT,
    category TEXT,
    price REAL
) STRICT;

-- Insert at least 3 items
INSERT INTO sports_gear (gear_name, category, price) VALUES
('Football Helmet', 'Equipment', 150.00),
('Team Hoodie', 'Apparel', 65.00),
('Water Bottle', 'Accessories', 15.00);
```

```python
%%sql
-- Show all items that cost more than $50
SELECT *
FROM sports_gear
WHERE price > 50;
```

```python
%%sql
-- Update a price
UPDATE sports_gear
SET price = 129.99
WHERE gear_name = 'Football Helmet';

-- Delete an item
DELETE FROM sports_gear
WHERE gear_name = 'Water Bottle';
```
